```python
# Divide Data into Test and Training Set
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(data, test_size = 0.2, random_state=42)
```

Obviously, the amount of Streams correlates greatly with the amount of views likes and comments on spotiy. But dont want to use those attributes but rather all other attributes.

```python
data = train_set.drop("Stream", axis=1)
data_label = train_set[["Stream"]].copy()
```

```python
#drop unnecessary columns first, then deal with naas
data_removed = data.drop(['Url_spotify', 'Track', 'Album', 'Album_type', 'Uri', 'Url_youtube', 'Channel', 'Views', 'Likes', 'Comments','Description', 'official_video', 'Title'], axis=1)
```

```python
data_removed[data_removed.isnull().any(axis=1)]
```

```python
#if title is Title is Naan, we want to drop the whole row
data_removed.dropna(subset=['Licensed'], inplace=True)
```

```python
data_removed[data_removed.isnull().any(axis=1)]
```

The key attribute describes the key that the song is written in. Because this is a categorical variable, i think it makes sense to convert it using oneHotEncoder

```python
key_cat = data_removed["Key"]
key_cat.head(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np

cat_encoder = OneHotEncoder()

key_cat_reshaped = key_cat.to_numpy().reshape(-1, 1)

key_cat_1hot = cat_encoder.fit_transform(key_cat_reshaped)
                                             
key_cat_1hot_df = pd.DataFrame(key_cat_1hot.toarray(), columns=cat_encoder.categories_)

train_set_encoded = pd.concat([data_removed, key_cat_1hot_df], axis=1)

# drop the original categorical variable from the encoded training set
train_set_encoded = train_set_encoded.drop('Key', axis=1)

pd.set_option("display.max_columns", None)

train_set_encoded.head()                              
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_reg = LinearRegression()
lin_reg.fit(data_removed, data_label)


X_test = test_set.drop("Stream", axis=1)
y_test = test_set[["Stream"]].copy() 
X_test_removed = X_test.drop(['Url_spotify', 'Track', 'Album', 'Album_type', 'Uri', 'Url_youtube', 'Channel', 'Views', 'Likes', 'Comments','Description', 'official_video', 'Title'], axis=1)
X_test_removed.dropna(subset=['Title'], inplace=True)


y_pred = lin_reg.predict(X_test_removed)


mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
